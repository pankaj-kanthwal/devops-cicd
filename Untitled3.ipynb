```
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
# %matplotlib inline

#importing the file
import pandas as pd
import io
from google.colab import files
uploaded = files.upload()

```

```
df = pd.read_csv(io.StringIO(uploaded['Default_On_Payment.csv'].decode('utf-8')))

import pandas_profiling

German_credit_pp = pandas_profiling.ProfileReport(df)

German_credit_pp.to_file("German_credit_profiling1.html")

German_credit_pp

#converting the categorial to numerical
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.Categorical(df[col]).codes

df.head()

```

```
#checking for missing values
pd.isnull(df).any()
pd.isnull(df).sum()

#dropping the missing values as they are really less in number
df = df.dropna(how='any')

#REchecking for missing values
pd.isnull(df).any()
pd.isnull(df).sum()

```

```
#Splitting into features and labels (x =independant , y = dependant)
x1 = df.drop('Default_On_Payment',axis =1)
x = x1.drop('Customer_ID',axis =1)

x.head(5)

y = df['Default_On_Payment']

y.head()
```

```
#test train split 
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size =0.30, random_state =1 )

train_x.shape
test_x.shape
train_y.shape
test_y.shape

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

from sklearn import preprocessing
from sklearn import metrics
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(train_x,train_y)
classes = dt.predict(test_x)
auc = metrics.roc_auc_score(test_y,classes)
auc

from sklearn.metrics import roc_curve, auc, recall_score,precision_score,accuracy_score,f1_score
from sklearn.metrics import confusion_matrix,average_precision_score,recall_score
```

```
# %matplotlib inline

dt = DecisionTreeClassifier(criterion='entropy', max_depth = 6)
dt.fit(train_x,train_y)
classes = dt.predict(test_x)
auc = metrics.roc_auc_score(test_y,classes)
auc

from sklearn.externals.six import StringIO  

import graphviz
import pydotplus

dot_data = tree.export_graphviz(dt, out_file=None, feature_names=x.columns, class_names='Default_On_Payment',filled=True,rounded=True,special_characters=True)

graph = graphviz.Source(dot_data)

graph
```
